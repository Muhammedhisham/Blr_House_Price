```python
from sklearn import preprocessing
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
data['area_type']=label_encoder.fit_transform(data['area_type'])
data['location']=label_encoder.fit_transform(data['location'])
data['society']=label_encoder.fit_transform(data['society'])
data['size']=label_encoder.fit_transform(data['size'])
data['availability']=label_encoder.fit_transform(data['availability'])
```

```python
sqt=[]
for i in data['total_sqft']:
    if '-' in i:
        
        k=i.split(' - ')
        k=(float(k[0])+float(k[1]))/2
       
        sqt.append(k)
    elif i.isnumeric():
        sqt.append(i)
    else:
        sqt.append(None)
data['total_sqft']=sqt
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
y=data['price']
x=data.drop('price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
x_test.shape
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
y_test
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model = lr.fit(x_train,y_train)
prediction=model.predict(x_test)
prediction
```

```python
model.score(x_test,y_test)

```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
print('MSE Is:',mean_squared_error(y_test,prediction))
print('R squared value is :',r2_score(y_test,prediction))
print('MAE value is :',mean_absolute_error(y_test,prediction))
```
